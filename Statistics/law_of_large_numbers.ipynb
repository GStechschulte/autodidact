## The Law of Large Numbers (LLN)

The LLN is not the most important in science. . .it is necessary the LLN holds in order to do statistical science

 - Assume $X_1. . .,X_n$ are random variables (RVs)
 - Assume i.i.d

The sample everage $\bar{X_n} = n^{-1}\sum_{i=1}^nX_i$ converges in probability to the expectation $\mu = E(X_i)$
 - This means that $\bar{X_n}$ is close to $\mu$ with high probability
 - The distribution of $\bar{X_n}$ (the sample) becomes more concentrated around $\mu$ (population) as $n$ gets large


The average will be more stable than the individual observations (more is different)
 - The mean converges to the expected value as more and more observations are recorded
 - "A portfolio of stocks will be more stable than just holding one stock"

$X_1$ = {0, 1} --> p = 0.5

$X_2$ = {0, 1} --> p = 0.5

($X_1 + X_2$) = $\frac{\sqrt{p(1-p)}}{\sqrt{2}}$

But what about thin tailed distributions such as the Gaussian vs Fat tailed distributions?

```python
population = np.random.gamma(1, 100, 10**3)
sns.histplot(population)
plt.title('Mean = {}'.format(round(population.mean(), 2)))
```

```python
# Sample from the population: n = ??
sample1 = np.random.choice(population, 5)
sample2 = np.random.choice(population, 100)
sample3 = np.random.choice(population, 250)

np.mean(sample1), np.mean(sample2), np.mean(sample3)
```
