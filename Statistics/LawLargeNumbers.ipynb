## The Law of Large Numbers (LLN)

The LLN is not the most important in science. . .it is necessary the LLN holds in order to do statistical science

 - Assume $X_1. . .,X_n$ are random variables (RVs)
 - Assume same i.i.d

The average will be more stable than the individual observations
 - The mean converges to the expected value as more and more observations are recorded
 - "A portfolio of stocks will be more stable than just holding one stock"

$X_1$ = {0, 1} --> p = 0.5

$X_2$ = {0, 1} --> p = 0.5

($X_1 + X_2$) = $\frac{\sqrt{p(1-p)}}{\sqrt{2}}$
